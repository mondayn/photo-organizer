```python
# goal: dlib does a good job finding faces, but I came across a couple of examples where it did not detect 
# faces and after opening these in pix and saving with enhanced quality, dlib detects their faces.  
# So I'll test enhancing to improve recognition 
# and see if reducing pic size reduces processing time without sacrificing accurancy.

from fx import *
```

```python
problem_pic='labeled_pics/IMG_0025-2017-12-26 141137.JPG'
# labeled_pics/IMG_0210-2016-11-06 211542.JPG  # works as well
# show(problem_pic)
print('faces found: ',len(get_encodings(problem_pic)))
```

```python
#try enhancing
img = cv2.imread(problem_pic)

# # denoise
img = cv2.fastNlMeansDenoisingColored(img, None, 10, 10, 7, 21)

# # normalize lighting
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab)

clahe = cv2.createCLAHE(clipLimit=3.0)
l = clahe.apply(l)

img = cv2.merge((l, a, b))
img = cv2.cvtColor(img, cv2.COLOR_LAB2BGR)

# # sharpen
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
img = cv2.filter2D(img, -1, kernel)

# cv2.imwrite(enhanced_pic, img)

print(f'faces found: ',len(face_encodings(img)))
```
